```python
tracts=demo20.GEOCODE
tracts_df=[]

for tract in tracts:
    tract_count=tract_count+1
    link=f"https://census-prod.news-engineering.aws.wapo.pub/{tract}.json"
    response=requests.get(link)
    data=response.json() 
    tract_d=data[f'{tract}']

    tract90=pd.DataFrame(tract_d["1990"], index=[0])
    tract90["tract_id"]=f'{tract}'
    tract90=tract90.rename(columns={'totpop':"totpop90",'hisp':"hisp90",'white':"white90", 'black':"black90", 
                                    'nativam':"nativam90",'asianpi':"asianpi90",'multi':"multi90",'diverse':"diverse90"})

    tract00=pd.DataFrame(tract_d["2000"], index=[0])
    tract00["tract_id"]=f'{tract}'
    tract00=tract00.rename(columns={'totpop':"totpop00", 'hisp':"hisp00", 'white':"white00", 'black':"black00", 
                                    'nativam':"nativam00",'asianpi':"asianpi00",'multi':"multi00",'diverse':"diverse00"})

    tract10=pd.DataFrame(tract_d["2010"], index=[0])
    tract10["tract_id"]=f'{tract}'
    tract10=tract10.rename(columns={'totpop':"totpop10", 'hisp':"hisp10", 'white':"white10", 'black':"black10", 
                                    'nativam':"nativam10",'asianpi':"asianpi10",'multi':"multi10",'diverse':"diverse10"})

    tract20=pd.DataFrame(tract_d["2020"], index=[0])
    tract20["tract_id"]=f'{tract}'
    tract20=tract20.rename(columns={'totpop':"totpop20", 'hisp':"hisp20", 'white':"white20", 'black':"black20", 
                                    'nativam':"nativam20",'asianpi':"asianpi20",'multi':"multi20",'diverse':"diverse20"})

    tract90=tract90[['tract_id', 'totpop90', 'hisp90', 'white90', 'black90',
           'nativam90', 'asianpi90', 'multi90', 'diverse90']]
    tract00=tract00[['tract_id', 'totpop00', 'hisp00', 'white00', 'black00',
           'nativam00', 'asianpi00', 'multi00', 'diverse00']]
    tract10=tract10[['tract_id', 'totpop10', 'hisp10', 'white10', 'black10',
           'nativam10', 'asianpi10', 'multi10', 'diverse10']]

    tract10_90=pd.merge(tract10,pd.merge(tract00, tract90))

    merged_tracts=pd.merge(tract20,tract10_90)
    
    tracts_df.append(merged_tracts)
final_df=pd.DataFrame(pd.concat(tracts_df, ignore_index=True))
```
