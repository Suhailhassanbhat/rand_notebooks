```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("In-Person Voting.csv")
del df["Unnamed: 18"]
del df["Unnamed: 19"]
```

```python
new_df=df[df["Date"] == "11/2/20"]
```

```python
new_df=new_df.drop(["All Ballots RETURNED - Daily", "All In-Person - Daily", "DEM In-Person - Daily","REP In-Person - Daily", "Date"], axis=1)
```

```python
new_df
```

```python
new_df["total_ballots"] = new_df["All Ballots RETURNED"] + new_df["All In-Person"]
new_df["total_dem_ballots"] = new_df["DEM Ballots RETURNED"]+new_df["DEM In-Person"]
new_df["total_rep_ballots"] = new_df["REP Ballots RETURNED"]+new_df["REP In-Person"]

new_df
```

```python
# In Person percentages

new_df["dems_in_person_pct"] = new_df["DEM In-Person"] *100/ new_df["total_dem_ballots"]
new_df["dems_in_person_pct"] =new_df["dems_in_person_pct"].round()

new_df["rep_in_person_pct"] = new_df["REP In-Person"] *100/ new_df["total_rep_ballots"]
new_df["rep_in_person_pct"] =new_df["rep_in_person_pct"].round()
```

```python
# ballots returned

new_df["ballotsReturned_total_pct"] = new_df["All Ballots RETURNED"] *100/new_df["All Ballots SENT"]
new_df["ballotsReturned_total_pct"] = new_df["ballotsReturned_total_pct"].round()

new_df["ballotsReturned_dem_pct"] = new_df["DEM Ballots RETURNED"] *100/new_df["DEM Ballots SENT"]
new_df["ballotsReturned_dem_pct"] = new_df["ballotsReturned_dem_pct"].round()

new_df["ballotsReturned_rep_pct"] = new_df["REP Ballots RETURNED"] *100/new_df["REP Ballots SENT"]
new_df["ballotsReturned_rep_pct"] = new_df["ballotsReturned_rep_pct"].round()
```

```python
filtered_df=new_df[["County",'total_ballots', 'total_dem_ballots',
       'total_rep_ballots', 'dems_in_person_pct', 'rep_in_person_pct',
       'ballotsReturned_total_pct', 'ballotsReturned_dem_pct',
       'ballotsReturned_rep_pct']].reset_index(drop=True)
```

```python
filtered_df.head()
```

```python
#import voter registration data
voter_reg =pd.read_csv("voter_registration.csv")

voter_reg["non_party"] = voter_reg["Other"] + voter_reg["Ind"] + voter_reg["Libert"] + voter_reg["Green"]+ voter_reg["Const"] + voter_reg["Reform"] + voter_reg["Soc Wk"] 

voter_reg = voter_reg[["County", 'Dem', 'Rep', "non_party", "Registered"]]
```

```python
merged_df = pd.merge(filtered_df, voter_reg, left_on="County", right_on="County")
```

```python
merged_df["early_vote_registered_pct"] = (merged_df["total_ballots"]*100/merged_df["Registered"]).round()
merged_df["early_voter_dem_registered_pct"] = (merged_df["total_dem_ballots"]*100/merged_df["Dem"]).round()
merged_df["early_voter_rep_registered_pct"] = (merged_df["total_rep_ballots"]*100/merged_df["Rep"]).round()
```
