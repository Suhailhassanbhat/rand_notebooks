```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
import os
import re
import camelot

# import ssl
# ssl._create_default_https_context = ssl._create_unverified_context
```

```python
#   ,"August","September","October","November","December"]
months_2021 = ["July", "June", "May", "April","March", "Feb.", "Jan."]
months_2020 = ["Dec.", 'Nov.', 'Oct.', 'Sept.', 'Aug.', 'July', 'June', 'May', 'April', 'March']
```

```python
page_list=[]
for month in months_2021:
    page_dict={}
    page_dict['month']= month + ", 2021"
    page_dict['page_link']=f'https://chfs.ky.gov/Pages/cvdaily.aspx?View={month}%202021%20Daily%20Summaries&Title=Table%20Viewer%20Webpart'
    
    page_list.append(page_dict)
for month in months_2020: 
    page_dict_2020 ={}
    page_dict_2020['month']=  month + ", 2020"
    page_dict_2020['page_link']=f'https://chfs.ky.gov/Pages/cvdailyarchive.aspx?View={month}%202020%20Daily%20Summaries&Title=Table%20Viewer%20Webpart'
    page_list.append(page_dict_2020)
pages_df=pd.DataFrame(page_list)

pages_df.to_csv("ky_monthly_covid_pages.csv", index=False)
```

```python
link_list=[]
for url in pages_df.page_link:
#     print(url)
    response=requests.get(url, verify=False)
    doc=BeautifulSoup(response.text, 'html.parser')
    container=doc.select_one('tbody').select('tr')
    for element in container:
        link_dict={}
        for ele in element.find_all('a'): 
            try: 
                link_dict['date']=ele.text.replace("COVID-19 Daily Report", "")
                link_dict['link']=ele.get('href')
            except:
                pass
        link_list.append(link_dict)
```

```python
df=pd.DataFrame(link_list)
df.date=df.date.str.replace("COVID19 Daily Report", "").str.replace("COVID 19 Daily Report", "").str.replace(" COVID-18 Daily Report", "")

df
```
