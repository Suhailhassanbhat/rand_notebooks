## BLOCK DATA

```python
#read 1990 data
df90=pd.read_csv("ky_block1990.csv", dtype={"STATEA":str,"COUNTYA":str,
                                            "BLCK_GRPA":str,"TRACTA":str,"BLOCKA":str })
```

```python
jeff90=df90[df90.COUNTYA == "111"].reset_index(drop=True)
```

```python
jeff90["GEOCODE90"]=jeff90.STATEA+jeff90.COUNTYA+jeff90.TRACTA+"00"+jeff90.BLOCKA
```

```python
jeff90=jeff90[['GISJOIN', 'GEOCODE90','ET2001', 'ET2002', 'ET2003', 'ET2004', 'ET2005',
       'ET2006', 'ET2007', 'ET2008', 'ET2009', 'ET2010']]
```

```python
#read tt=otal population here
total90=pd.read_csv("block2019_totalpop.csv", dtype={"STATEA":str,"COUNTYA":str,
                                            "BLCK_GRPA":str,"TRACTA":str,"BLOCKA":str })
total90=total90[total90.COUNTYA == "111"].reset_index(drop=True)
total90=total90[["GISJOIN", "ET1001"]]
```

```python
merged90=pd.merge(jeff90,total90)
```

```python
merged90=merged90.rename(columns={"ET1001":"pop90",
                                     "ET2001":"white90",
                                     "ET2002":"black90",
                                     "ET2003":"aiea90",
                                     "ET2004":"api90",
                                     "ET2005":"otherr",
                                     "ET2006":"hwhite90",
                                     "ET2007":"hblack90",
                                     "ET2008":"haiea90",
                                     "ET2009":"hapi90",
                                     "ET2010":"hotherrace90"})
```

```python
merged90["other_race90"]=merged90.aiea90+merged90.api90+merged90.otherr
merged90["hispanic90"]=merged90.hwhite90+merged90.hblack90+merged90.haiea90+merged90.hapi90+merged90.hotherrace90
```

```python
merged90=merged90[['GISJOIN', 'GEOCODE90',"white90", "black90","hispanic90", "other_race90","pop90"]]
```

```python
merged90[['white90', 'black90', 'hispanic90',
       'other_race90', 'pop90']]=merged90[['white90', 'black90', 'hispanic90',
       'other_race90', 'pop90']].apply(pd.to_numeric)
```
